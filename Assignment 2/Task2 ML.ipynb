```python
# In the dataset crx.data, name the columns as col1, col2..col16

import pandas as pd
import numpy as np

#Load the data
data = pd.read_csv('crx.data', header=None)

#Variable engineering or feature
#col1 col2 col3...col16
varnames =['col'+str(s) for s in range(1,17)]

#add column names
data.columns = varnames
```

```python
# Print the last 10 rows of data

data.tail(10)
```

```python
# Replace the ‘?’ with Not-a-Number

data = data.replace('?',np.nan)
```

```python
# Comment on the datatype of variables

data.info()
```

```python
# The col16 has + and -, replace them ‘P’ and ‘N’ respectively

data['col16'] = data['col16'].map({'+':'P', '-':'N'})
```

```python
data.head(10)
```

```python
data.tail(10)
```

```python
# Find and display the number of variables of type ‘Object’

categorical_columns = [i for  i in data.columns if data[i].dtype == "O"]
categorical_columns

# Total object type of data is 12 as mention below.
```

## II Dataset: loan.csv
## Tasks:
#### 1. Display the mean of any two variables with continuous values
#### 2. Print the number of discrete variables
#### 3. Display the unique values of two variables with discrete values
#### 4. Display the Month with most of loans issued date
#### 5. Display the count of ‘Teacher’ who are ‘Owners’
#### 6. Display the ‘Employment’ of customers who mostly ‘Rent’

```python
# Import Libraries 
import pandas as pd
import numpy as np

# Load the data
data = pd.read_csv('loan.csv')
data.head()
```

```python
# Display the mean of any two variables with continuous values

data[['disbursed_amount','interest']].mean()
```

```python
# Print the number of discrete variables

Discrete_columns = [i for  i in data.columns if data[i].dtype == "O"]
Discrete_columns
```

```python
# Display the unique values of two variables with discrete values

data['market'].unique()
```

```python
data['employment'].unique()
```

```python
# Display the Month with most of loans issued date

data['date_issued']=pd.to_datetime(data['date_issued'])
month = data['date_issued'].dt.month
month.value_counts()
```

```python
# Display the count of ‘Teacher’ who are ‘Owners’

df = data[["employment",'householder']]
df = df.loc[df['employment'] == 'Teacher']
df = df.loc[df['householder'] == 'OWNER']
df.value_counts()
```

```python
# Display the ‘Employment’ of customers who mostly ‘Rent’

df1 = data[["employment",'householder']]
df1 = df1.loc[df1['householder'] == 'RENT']
df1.value_counts()
```
